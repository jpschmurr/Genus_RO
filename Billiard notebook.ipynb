```python
import networkx as nx
import math
def make_billiard_graph_integers(bracket):
  '''
"bracket" is a list of integers. They,together with n, should be relatively prime -- that is, they should not all share the same divisor.
If this rule is followed then the output should be a connected k-regular graph (expressed as a NetworkX object), where k
is the length of the bracket.
  '''
  B=nx.Graph() #empty graph
  #We assume that an admissible k-tuple is being input.
  k=len(bracket)
  n=sum(bracket)//(k-2)
  for i in range(2*n): #adds nodes to graph
    B.add_node(i)
  for rotation_vertex in range(n):#adds edges
    ref_list=[0]
    for p in range(1,k):
      ref_list.append((ref_list[p-1]+n-bracket[p])%n)
    for a in ref_list:
      flip_vertex=((a+rotation_vertex)%n)+n
      B.add_edge(rotation_vertex,flip_vertex)
  return B
```

```python
#Good list:[1,2,4],[2,3,5],[9,10,11],[7,8,15]
#Also try: [1,1,1,1], [1,2,3,8],[1,2,3,10],[2,5,8,9],[1,2,3,2n] for n>=4
#Also:[3,3,3,3,3],[1,2,3,4,14],[1,2,3,4,2+3n], for n>=0
#Also: [1,2,3,4,5,19],[1,2,3,4,5,27]
bracket=[1,2,3,8]
G=make_billiard_graph_integers(bracket)
nx.draw(G,with_labels=True)
list(G.nodes)
list(G.edges)
```
